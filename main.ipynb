```python
# pip install gspread
# pip install oauth2client

import pandas as pd
import os
import sys
sys.path.append('dir/')
import whatsapp
from functions import get_sheet, validate_worksheet_names, get_last_class_date, check_mobile_number, get_st_info



def main():
    
    url = 'https://docs.google.com/spreadsheets/d/1xl9ol0ieysuTjvbJIgXg_37iLGEWMU_jae5TE284Eng/edit#gid=762039304'
    full_sheet = get_sheet(url)
    
    # get name of all worksheets of the spread sheet
    actual_worksheet_names = full_sheet.worksheets()
    
    # read worksheet_names.txt file for given worksheet names
    with open('worksheet_names.txt', 'r') as f:
        wsl = f.read().split(',')
        given_worksheet_names = [i.strip() for i in wsl]

    # validate given 
    validated_ws_names = validate_worksheet_names(actual_worksheet_names, given_worksheet_names)


    for ws_name in validated_ws_names:  
        # worksheet = sheet.get_worksheet(0) 
        # Assuming you want to access the first worksheet (index 0)
        worksheet = full_sheet.worksheet(ws_name)
        data = worksheet.get_all_values()
        df = pd.DataFrame(data)
        student_info = get_st_info(df)
    

    
        # read message from text file
        with open('message.txt', 'r') as f:
            message_body = f.read()

        # creating list of numbers to send messages
        msg_to_be_sent = []
        for item in student_info:
            if item['Status'] != 'TRUE':
                msg_to_be_sent.append(item['Mobile No'])


        # sending message using send_message function from whatsapp module        
        for nums in msg_to_be_sent:
            for num in nums:
                whatsapp.send_message('01922444626', message_body)
                break
            break
        
        
        break


if __name__ == '__main__':
    main()
```
